## Analysis

```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt
```

1. Load the downloaded PDB file into MDAnalysis

```python
pdb_file = "/Users/user/Downloads/1gjz.pdb"
protein = mda.Universe(pdb_file)
```

2. Calculate the center of mass of the protein

```python
center_of_mass = protein.atoms.center_of_mass()
print(center_of_mass)
```

3. Calculate the distance between specific atoms

```python
atom1 = protein.select_atoms("resid 1 and name CA")
atom2 = protein.select_atoms("resid 10 and name CB")
distance = mda.lib.distances.distance_array(atom1.positions, atom2.positions)[0]
```

4. Analyze the secondary structure

```python
#for segment IDs
protein.atoms.segments.segids
print("Segment IDs:", protein.atoms.segments.segids)
```

5. Visualize the protein structure

## Visualization

```python
# Get coordinates of all atoms
coordinates = protein.atoms.positions
```

```python
# Extract x, y, and z coordinates
x = coordinates[:, 0]
y = coordinates[:, 1]
z = coordinates[:, 2]
```

```python
# Create a 3D scatter plot of the protein structure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', marker='o', s=20)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
